```python
import pandas as pd
from flask import Markup
games_info = pd.read_csv("game_id_cat.txt")
```

```python
games_info.iloc[4].desc_1
```

```python


def first(y):
    
    x = Markup(y).unescape()
    
    if "descr" in x.split("\n\n",1)[0].lower():
        return x.split("\n\n",2)[0] + "\n\n" + x.split("\n\n",2)[1]
    else:
        return x.split("\n\n",1)[0]
        
def second(x):
    if "descr" in x.split("&#10;&#10;",1)[0].lower():
        try:
            return x.split("&#10;&#10;",2)[2]
        except:
            return ""
            
    else:
        try:
            return x.split("&#10;&#10;",1)[1]
        except:
            return ""
    
```

```python
games_info["desc_1"] = games_info.description.apply(lambda x: first(str(x)))
games_info["desc_2"] = games_info.description.apply(lambda x: second(str(x)))
```

```python
print(games_info.iloc[7].desc_1)
```

```python
print(Markup(games_info.iloc[7].desc_2).unescape())
```

```python
games_info.drop("description", axis = 1, inplace=True)
```

```python
games_info
```

```python
games_info.to_csv("game_id_cat.txt")
```
